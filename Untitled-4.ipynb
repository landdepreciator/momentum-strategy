```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

# Download data
data = yf.download('AES', start="2015-01-01", end="2025-01-01")
if "Adj Close" in data.columns:
    data["Close"] = data["Adj Close"]
else:
    data["Close"] = data["Close"]

# Backtesting function
def backtest_strategy(data, short_window, long_window):
    data["SMA_short"] = data["Close"].rolling(window=short_window).mean()
    data["SMA_long"] = data["Close"].rolling(window=long_window).mean()

    # Generate signals
    data["Signal"] = 0
    data.loc[data["SMA_short"] > data["SMA_long"], "Signal"] = 1
    data["Position"] = data["Signal"].shift(1)

    # Strategy returns
    data["Daily_Return"] = data["Close"].pct_change()
    data["Strategy_Return"] = data["Position"] * data["Daily_Return"]

    # Cumulative returns
    data["Cumulative_Strategy"] = (1 + data["Strategy_Return"]).cumprod()
    data["Cumulative_Benchmark"] = (1 + data["Daily_Return"]).cumprod()

    # Performance metrics
    total_return_strategy = data["Cumulative_Strategy"].iloc[-1] - 1
    total_return_benchmark = data["Cumulative_Benchmark"].iloc[-1] - 1
    max_drawdown = ((data["Cumulative_Strategy"] / data["Cumulative_Strategy"].cummax()) - 1).min()

    # Sharpe ratio calculation (assuming risk-free rate is 0 for simplicity)
    sharpe_ratio = data["Strategy_Return"].mean() / data["Strategy_Return"].std() * np.sqrt(252)  # Annualized Sharpe ratio

    return total_return_strategy, total_return_benchmark, max_drawdown, sharpe_ratio

# Random search for optimization
best_params = None
best_performance = -np.inf
results = []

random.seed(42)
num_iterations = 1000  # Number of random parameter combinations

for _ in range(num_iterations):
    short_window = random.randint(5, 50)  # Random short window
    long_window = random.randint(51, 200)  # Random long window
    if short_window >= long_window:
        continue  # Skip invalid combinations

    total_return_strategy, total_return_benchmark, max_drawdown, sharpe_ratio = backtest_strategy(data.copy(), short_window, long_window)
    results.append((short_window, long_window, total_return_strategy, max_drawdown, sharpe_ratio))

    # Optimize based on Total Return
    performance_metric = total_return_strategy
    if performance_metric > best_performance:
        best_performance = performance_metric
        best_params = (short_window, long_window)

print(f"Best Parameters Found: SMA_short={best_params[0]}, SMA_long={best_params[1]}")
print(f"Best Performance (Total Return): {best_performance:.2%}")

total_return_strategy, total_return_benchmark, max_drawdown, sharpe_ratio = backtest_strategy(data.copy(), best_params[0], best_params[1])
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Maximum Drawdown: {max_drawdown:.2%}")

# Visualize the optimized strategy
data = data.copy()
backtest_strategy(data, best_params[0], best_params[1])

plt.figure(figsize=(12, 6))
plt.plot(data["Cumulative_Strategy"], label="Optimized Strategy", linewidth=2)
plt.plot(data["Cumulative_Benchmark"], label="Benchmark (Buy & Hold)", linewidth=2)
plt.title(f"Optimized Strategy vs Benchmark\n"
          f"Total Return: {data['Cumulative_Strategy'].iloc[-1] - 1:.2%} (Strategy), "
          f"{data['Cumulative_Benchmark'].iloc[-1] - 1:.2%} (Benchmark)")
plt.xlabel("Date")
plt.ylabel("Cumulative Returns")
plt.legend()
plt.grid()
plt.show()
```
